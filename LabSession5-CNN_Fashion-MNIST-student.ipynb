First, download and read the Fashion-MNIST data.

```python
%matplotlib inline
import keras
from keras.datasets import fashion_mnist

(x_trainD, y_train), (x_testD, y_test) = fashion_mnist.load_data()
```

### Part 1 - Data management and visualization

### Part 2 - Multi-layer perceptrons

#### 3) Define a simple Keras multi-layer perceptron (MLP) architecture using Sequential API and containing 3 dense layers: 2 dense layers with 512 units and ReLU activation followed by 1 dense layer for softmax regression. Use your model to get first classification results on Fashion MNIST.

```python
num_classes = D
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import RMSprop
```

```python
#data formating

x_train = x_trainD.astype(float)/255
x_test = x_testD.astype(float)/255


z_train=keras.utils.to_categorical(y_train,num_classes)
z_test=keras.utils.to_categorical(y_test,num_classes)

x_train_flat = x_train.reshape(A,C*C)
x_test_flat = x_test.reshape(B,C*C)
```

```python
mlp_1 = Sequential()

mlp_1.add(Dense(units=512, input_shape=(C*C,),activation="relu"))
mlp_1.add(Dense(units=512,activation="relu"))
mlp_1.add(Dense(10, activation="softmax"))

mlp_1.summary()

mlp_1.compile(RMSprop(), 'categorical_crossentropy', metrics=['accuracy'])
```

```python
batch_size = 128
epochs = 20

hist_mlp_1 = mlp_1.fit(x_train_flat, z_train, validation_data=(x_test_flat,z_test), epochs=epochs, batch_size=batch_size, verbose=2)
```

#### 4) Define a generic function evaluate_model() able to:
- plot the evolution of accuracy for both training and testing data with respect to the epochs
- compute final test loss and accuracy. 

Use this function to assess the performance of the previously defined model.

```python
def evaluate_model(model,history,x_eval,z_eval):
    
    # list all data in history
    print(history.history.keys())

    #Visualize history (loss vs epochs)
    plt.figure()
    plt.subplot(121)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('model loss')
    plt.ylabel('loss')  
    plt.xlabel('epochs')
    plt.legend(['train','val'], loc='upper left')
    plt.subplot(122)
    plt.plot(history.history['acc'])
    plt.plot(history.history['val_acc'])
    plt.title('model accuracy')
    plt.ylabel('acc')  
    plt.xlabel('epochs')
    plt.legend(['train','val'], loc='upper left')
    plt.subplots_adjust(right=3,top=2)
    plt.show()
    
    score = model.evaluate(x_eval, z_eval, verbose=0)
    print('Test loss:', score[0])
    print('Test accuracy:', score[1])
    
```

```python
evaluate_model(mlp_1,hist_mlp_1,x_test_flat,z_test)
```

##### Comments: Accuracy of 88.66% is not bad but we see the generalisation error from 3~4 epochs

### Part 3 - Convolutional neural networks

##### 6) Define a simple Keras convolutional neural network (CNN) architecture using Sequential API and containing the following layers : 1 convolution layer using Conv2D and 32 filters, 1 max-polling layer using MaxPooling2D, 2 dense layers using ReLU and softmax activations respectively. Use this model to get classification results (use evaluate_model() as previously) and compared them to MLP results. 

```python
# build 4D tensors
x_train_cnn = x_train.reshape(A, C, C, 1)
x_test_cnn = x_test.reshape(B, C, C, 1)

input_shape = (C, C, 1)
```

```python
from keras.layers import Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras import backend as K
```

```python
cnn_1 = Sequential()

cnn_1.add(Conv2D(32, (3,3), activation="relu", input_shape = input_shape, padding="SAME"))
cnn_1.add(MaxPooling2D())

cnn_1.add(Flatten())

cnn_1.add(Dense(units=6272,activation="relu"))
cnn_1.add(Dense(units=num_classes, activation="softmax"))

cnn_1.summary()

cnn_1.compile(RMSprop(), 'categorical_crossentropy', metrics=['accuracy'])

```

```python
batch_size = 128
epochs = 12

hist_cnn_1 = cnn_1.fit(x_train_cnn, z_train, validation_data=(x_test_cnn,z_test), epochs=epochs, batch_size=batch_size, verbose=2)
evaluate_model(cnn_1,hist_cnn_1,x_test_cnn,z_test)
```

###### Comments: The accuracy (91.4%) is better than the MLPs ones.    There is a lot of paramaters (almost 40millions because of the fully connected layer), we do overfitting quickly as we see the generalization error early on

#### 7) As for MLP, define a deeper CNN architecture to outperform the previously obtained classification results. What are your conclusions?

```python
cnn_2 = Sequential()

cnn_2.add(Conv2D(64, (5, 5), input_shape=input_shape, padding='same',activation='relu'))
cnn_2.add(MaxPooling2D(pool_size=(2, 2),padding='same'))

cnn_2.add(Dropout(0.3))

cnn_2.add(Conv2D(64, (3, 3), padding='same',activation='relu'))
cnn_2.add(MaxPooling2D(pool_size=(2, 2),padding='same'))

cnn_2.add(Dropout(0.3))

cnn_2.add(Flatten())
    
cnn_2.add(Dense(256, activation='relu'))

cnn_2.add(Dropout(0.5))

cnn_2.add(Dense(64, activation='relu'))
    
cnn_2.add(Dense(num_classes, activation='softmax'))

cnn_2.compile(RMSprop(),'categorical_crossentropy', metrics=['accuracy'])

cnn_2.summary()
```

#### 8) Create a function visualize_wrongly_predictions() able to show wrong prediction for each class for a given model as input. Use this function for each of the 4 previouly trained models (2 MLP, 2 CNN).

#### 9) Using the code given below, create a function visualize_confusion_matrix() able to display the confusion matrix  for a given model as input. Use this function for each of the 4 previouly trained models (2 MLP, 2 CNN). What is the most common confusion?

```python
from sklearn.metrics import confusion_matrix
import itertools

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        #print("Normalized confusion matrix")
    #else:
        #print('Confusion matrix, without normalization')

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
def visualize_confusion_matrix(model,x_test,y_test,title):
    z_pred= model.predict(x_test)
    y_pred=[np.argmax(i) for i in z_pred]
    cm = confusion_matrix(y_test, y_pred)
    np.set_printoptions(precision=2)
    plt.figure()
    plot_confusion_matrix(cm, class_names,normalize=True, title=title)
    
```

```python
visualize_confusion_matrix(mlp_1,x_test_flat,y_test,"Normalized Confusion matrix of mlp_1")
visualize_confusion_matrix(mlp_2,x_test_flat,y_test,"Normalized Confusion matrix of mlp_2")
```

```python
visualize_confusion_matrix(cnn_1,x_test_cnn,y_test,"Normalized Confusion matrix of cnn_1")
visualize_confusion_matrix(cnn_2,x_test_cnn,y_test,"Normalized Confusion matrix of cnn_2")
```

#### 10) Based on the documentation (https://keras.io/preprocessing/image/), train one of your CNN architecture using data augmentation and conclude on the used of augmented data.

Transfer learning, is a research problem in machine learning that focuses on storing knowledge gained while solving one problem and applying it to a different but related problem.

In this part, you will do fine tuning on a pre-trained network. Fine-tuning consists in starting from a trained network, then re-training it on a new dataset using very small weight updates. To do so, you need to load a pre-trained model, add a dense classifier to compute the output, and then to freeze the weights of the pre-trained model (https://blog.keras.io/building-powerful-image-classification-models-using-very-little-data.html).   

```python
from keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rotation_range=8,
    width_shift_range=0.08,
    height_shift_range=0.08,
    shear_range=0.3, 
    zoom_range=0.08)



data_aug = datagen.flow(x_train_cnn,z_train,batch_size=1)
```

```python
cnn_2 = Sequential()

cnn_2.add(Conv2D(64, (5, 5), input_shape=input_shape, padding='same',activation='relu'))
cnn_2.add(MaxPooling2D(pool_size=(2, 2),padding='same'))

cnn_2.add(Dropout(0.3))

cnn_2.add(Conv2D(64, (3, 3), padding='same',activation='relu'))
cnn_2.add(MaxPooling2D(pool_size=(2, 2),padding='same'))

cnn_2.add(Dropout(0.3))

cnn_2.add(Flatten())
    
cnn_2.add(Dense(256, activation='relu'))

cnn_2.add(Dropout(0.5))

cnn_2.add(Dense(64, activation='relu'))
    
cnn_2.add(Dense(num_classes, activation='softmax'))

cnn_2.compile(RMSprop(),'categorical_crossentropy', metrics=['accuracy'])

cnn_2.summary()
```

```python
hist2_cnn_2 = cnn_2.fit_generator(data_aug,steps_per_epoch=1875,validation_data=(x_test_cnn,z_test),epochs=12,verbose=2)
```

```python
evaluate_model(cnn_2,hist2_cnn_2,x_test_cnn,z_test)
```

**we create a model based on vgg16. We don't take the last last and add our own dense layer with 10 neurons**  
we set the internal parameters of the original network to not trainable and the last dense to trainable. 


```python
from keras.models import Model
from keras.optimizers import Adam

#We add the last layer in the vgg16
output=VGG16.output
output=Flatten(name='flatten')(output)
output=Dense(units=10,activation="softmax")(output)

modelvgg16=Model(inputs=VGG16.input,outputs=output)

#we freeze the weights except for the last layer (classification)
for layer in modelvgg16.layers[:-1]:
    layer.trainable = False
modelvgg16.layers[-1].trainable = True
modelvgg16.compile(Adam(),'categorical_crossentropy', metrics=['accuracy'])
modelvgg16.summary()
```

**However, the training does not seem to be working**  
The acc remains at 0.1 (as if it was a random decision)

```python
epochs=12
batch_size=32

hist_vgg16=modelvgg16.fit(X, z_train, validation_data=(X_test,z_test), epochs=epochs, batch_size=batch_size, verbose=2)
```

```python
evaluate_model(modelvgg16,hist_vgg16,X_test,z_test)
```
